# 'Fitting the Yield Curve

Q)Use either Nelson-Siegel, or Nelson-Siegel-Svensson to fit the Yield Curve.  Write out the equation.

the Nelson-Siegel, or Nelson-Siegel-Svensson is used to plot the zero coupen rates and implied forward rate for different yield to maturities 
the equation is given by
f (t) = β0 + β1 * exp(-t / Ʈ1) + β2 * exp(-t / Ʈ1) * (-t / Ʈ1) + β3 * exp(-t / Ʈ2) * (-t / Ʈ2)
and zero coupen rate is given by 
r(t) = β0 + (β1 + β2 ) * (1- exp-(t / Ʈ1))/( t /Ʈ1) — β2 * exp(-t / Ʈ1)) + β3 * (1- exp-(t / Ʈ2))/( t / Ʈ2) — β3 * exp(-t / Ʈ2))
β0, β1, β2 and Ʈ : model parameters that are to be estimated. These are parameters respectively for long term rate, short term rate, slope and degree of hump.
Ʈ2 and β3 explain the hump and level and direction of the hump in the curve

```python
# checking for stationarity an dimporting tools to plot acf and pacf 
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
# renaming columns as per question 
gold_etf_ts.rename(columns={'Adj Close':'price'},inplace=True) 
```

```python
gold_etf_ts.plot()
# we can see there is downword trend  so time series may not be statioanry 
```

```python
#Check for stationariy for series 
def adf_stationarity_test(price):
    indices= ['Test Statistic', 'p-value','no of Lags Used', 'no of Observations Used']
    adf_sta_test=adfuller(price,autolag="AIC")
    print(adf_sta_test)
    result=pd.Series(adf_sta_test[0:4],index=indices)
    for key,value in adf_sta_test[4].items():
        print(key)
        print(value)
        result[f'Critical value({key})']=value
    
    return result
adf_stationarity_test(gold_etf_ts.price)
#The null hypothesis of the ADF test states that the time series is not stationary. hence series is not stationary 
```

```python
# calculating first difference to make series statioanry 
gold_etf_ts_diff=gold_etf_ts.diff().dropna()
```

```python
# plots shwos time series stationary 
gold_etf_ts_diff.plot()
```

```python
gold_etf_ts_diff=gold_etf_ts_diff.dropna()
def adf_stationarity_test(price):
    indices= ['Test Statistic', 'p-value','no of Lags Used', 'no of Observations Used']
    adf_sta_test=adfuller(price,autolag="AIC")
    result=pd.Series(adf_sta_test[0:4],index=indices)
    for key,value in adf_sta_test[4].items():
        print(key)
        print(value)
        result[f'Critical value({key})']=value
    
    return result
adf_stationarity_test(gold_etf_ts_diff.price)
#The null hypothesis of the ADF test states that the time series is not stationary. now p value is less than 5% hence series is  stationary
```

```python
#plotting acf and pacf plots 

N_LAGS=40
SIGNIFICANCE_LEVEL = 0.05
fig, ax = plt.subplots(2, 1)
plot_acf(
gold_etf_ts_diff.price,ax=ax[0],lags=N_LAGS,alpha=SIGNIFICANCE_LEVEL
)
plot_pacf(
gold_etf_ts_diff.price,ax=ax[1],lags=N_LAGS,alpha=SIGNIFICANCE_LEVEL
)
```

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.diagnostic import acorr_ljungbox
import scipy.stats as scs

#ARIMA(0,1,1)
arima = ARIMA(gold_etf_ts, order=(0, 1, 1)).fit(disp=0)
arima.summary()
```

```python
#  analysing  where the model is good fit or not 
def arima_diagnostics(resids, n_lags=40):
 # create placeholder subplots
     fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
     r = resids
     resids = (r - np.nanmean(r)) / np.nanstd(r)
     resids_nonmissing = resids[~(np.isnan(resids))]
     # residuals over time
     sns.lineplot(x=np.arange(len(resids)), y=resids, ax=ax1)
     ax1.set_title('Standardized residuals')
     # distribution of residuals
     x_lim = (-1.96 * 2, 1.96 * 2)
     r_range = np.linspace(x_lim[0], x_lim[1])
     norm_pdf = scs.norm.pdf(r_range)
     sns.distplot(resids_nonmissing, hist=True, kde=True,
     norm_hist=True, ax=ax2)
     ax2.plot(r_range, norm_pdf, 'g', lw=2, label='N(0,1)')
     ax2.set_title('Distribution of standardized residuals')
     ax2.set_xlim(x_lim)
     ax2.legend()
     # Q-Q plot
     qq = sm.qqplot(resids_nonmissing, line='s', ax=ax3)
     ax3.set_title('Q-Q plot')
#     Time Series Modeling Chapter 3
#     [ 117 ]
     # ACF plot
     plot_acf(resids, ax=ax4, lags=n_lags, alpha=0.05)
     ax4.set_title('ACF plot')
     return fig
#Test the residuals of the fitted ARIM
arima_diagnostics(arima.resid, 40)
```

```python
#errors distribution is left skewed so model is not good fit 
```

```python
#8. Apply the Ljung-Box test for no autocorrelation in the residuals and plot the
#results:
ljung_box_results = acorr_ljungbox(arima.resid)
fig, ax = plt.subplots(1, figsize=[16, 5])
sns.scatterplot(x=range(len(ljung_box_results[1])),
 y=ljung_box_results[1],ax=ax)
ax.axhline(0.05, ls='--', c='r')
ax.set(title="Ljung-Box test's results",xlabel='Lag',ylabel='p-value')
```
