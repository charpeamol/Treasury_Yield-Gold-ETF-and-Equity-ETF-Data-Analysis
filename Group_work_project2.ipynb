# Assesing Stationary

we are using  Augmented Dickeyâ€“Fuller test to check stationary of all 3 series 
augmented Dickeyâ€“Fuller test (ADF) tests the null hypothesis that a unit root is present in a time series sample
H0: Time Series is not stationary
H1:time series is stationary 

# Checking stationarity of gold etf 

```python
from statsmodels.tsa.stattools import adfuller, kpss
def adf_stationarity_test(price):
    indices= ['Test Statistic', 'p-value','no of Lags Used', 'no of Observations Used']
    adf_sta_test=adfuller(price,autolag="AIC")
    print(adf_sta_test)
    result=pd.Series(adf_sta_test[0:4],index=indices)
    for key,value in adf_sta_test[4].items():
        print(key)
        print(value)
        result[f'Critical value({key})']=value
    
    return result
adf_stationarity_test(gold_etf.Close)
```

the null hypothesis of the Augmented Duckey Fuller  test states that the time series is not stationary.  
p value is 12% greater than 5% hence fail to reject null hypothesis hence time series GOLD_ETF not stationary

# Vector Error Correction model Q2

```python
from statsmodels.tsa.vector_ar.vecm import VECM
VECM_ETF_Q2 = VECM(endog = data_vecm_Q2[['bitcoin_Q2', 'gold_etf_Q2']], k_ar_diff=6, coint_rank=1, deterministic='cili',
                  seasons=12, first_season=1)
VECM_ETF_Q2_fit = VECM_ETF_Q2.fit()
VECM_ETF_Q2_fit.summary()
```

```python
irf = VECM_ETF_Q2_fit.irf(50)
irf.plot(orth=False)
```

# inferences from Vector Error Correction Model 
The loading coefficient (alpha) for GOLD_ETF is  not statistically significant, 
but Equity_ETF  is highly significant and estimated at -0.2011. 
This information means Gold_ETF is  weakly exogenous to Equity_ETF. 
The beta coefficient for Equity_ETF is, again, standardized at 1 for ease of interpretation of the beta coefficients for GOLD_ETF  which is -1.9364 . 
This means a 1% increase in Equity_ETF leads to a 1.93% decrease in Gold_ETF in the long-run and 20% of this correction occurs within a month.
